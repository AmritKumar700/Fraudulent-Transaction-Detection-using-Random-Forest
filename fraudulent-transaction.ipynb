# Model Selection & Training

```python
y = df['Class']
```

```python
x = df.drop(columns = ['Class'], axis = 1)
```

```python
x
```

```python
y
```

```python
# selection of model
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
# Divide dataset into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
rfc = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 32, max_depth = 10)
```

```python
rfc.fit(x_train, y_train)
```

# Model Evaluation | Prediction & Interpretation

```python
# Testing our model, and predict 
y_pred = rfc.predict(x_test)
```

```python
from sklearn.metrics import  classification_report, confusion_matrix
```

The classification_report (from sklearn.metrics) gives you a detailed summary of how well your classification model performed:

```python
print('Classifcation report \n', classification_report(y_test, y_pred))
```

```python
conf_mat = confusion_matrix(y_true = y_test, y_pred = y_pred)
print('Confusion matrix:\n', conf_mat)
```

```python
# Checking the accuracy of our model, how well our model performe
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

Trying with undersmapling technique - This is the pipeline module we need from imblearn for Undersampling

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```python
rand_sampling = RandomUnderSampler()
model = RandomForestClassifier(n_estimators = 30,criterion = 'entropy', random_state = 32, max_depth = 10)
```

```python
# Define the pipeline and combine sampling method with the RF model
pipeline = Pipeline([('RandomUnderSampler', rand_sampling), ('RF', model)])
```

```python
pipeline.fit(x_train, y_train) 
pred = pipeline.predict(x_test)
```

```python
# Result from the classification report and confusion matrix 
print('Classifcation report: ', classification_report(y_test, pred))
conf_mat = confusion_matrix(y_true = y_test, y_pred = pred)
print('Confusion matriX: ', conf_mat)
```
